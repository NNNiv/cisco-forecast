```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data_pack.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
def get_quarters():
    """Fetch the quarters in YYYY-MM-DD format"""

    quarters = list(df.columns[3:15])
    dates = {
        "Q1": "10-31",
        "Q2": "01-31",
        "Q3": "04-30",
        "Q4": "07-31",
    }

    return [f"20{q[2:4]}-{dates[q[-2:]]}" for q in quarters]
```

```python
for _, row in df.iterrows():
    quarters = [str(q) for q in get_quarters()]
    units = [0 if type(row[q]) != int else int(row[q]) for q in df.columns[3:15]]
    product_data = pd.DataFrame({
        "quarter": quarters,
        "units": units,
    })
    plt.figure(figsize=(10, 5))
    plt.plot(product_data["quarter"], product_data["units"], marker="o")
    plt.title(str(row["Product Name"]))  
    plt.ylabel("Units Sold")
    plt.xlabel("Quarter")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
